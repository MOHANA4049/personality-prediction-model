```python
import pandas as pd
from numpy import *
import numpy as np
from sklearn import preprocessing
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
from sklearn import neighbors
```

```python
#Input as training data
data =pd.read_csv('train dataset.csv')
array = data.values
```

```python
# Displaying training data
data
```

```python
#Preprocessing
#Converting Categorical to numerical value
for i in range(len(array)):
	if array[i][0]=="Male":
		array[i][0]=1
	else:
		array[i][0]=0
```

```python
#Result of conversion
df=pd.DataFrame(array)
df
```

```python
maindf =df[[0,1,2,3,4,5,6]]
mainarray=maindf.values
print (mainarray)
```

```python
temp=df[7]
train_y =temp.values
# print(train_y)
# print(mainarray)
train_y=temp.values
```

```python
for i in range(len(train_y)):
	train_y[i] =str(train_y[i])
```

```python
#Applying Logistic Regression
mul_lr = linear_model.LogisticRegression(multi_class='multinomial', solver='newton-cg',max_iter =1000)
#Training the data
mul_lr.fit(mainarray, train_y)
```

```python
#Getting test data
testdata =pd.read_csv('test dataset.csv')
test = testdata.values
testdata
```

```python
#Preprocessing
#Converting Categorical to numerical value
for i in range(len(test)):
	if test[i][0]=="Male":
		test[i][0]=1
	else:
		test[i][0]=0
```

```python
df1=pd.DataFrame(test)
```

```python
testdf =df1[[0,1,2,3,4,5,6]]
maintestarray=testdf.values
print(maintestarray)
```

```python
#Predicting Personality
y_pred = mul_lr.predict(maintestarray)
for i in range(len(y_pred)) :
	y_pred[i]=str((y_pred[i]))
	print("Personality for Person No "+str(i)+" is " + y_pred[i])
DF = pd.DataFrame(y_pred,columns=['Predicted Personality'])
DF.index=DF.index+1
DF.index.names = ['Person No']
DF.to_csv("output.csv")
```
